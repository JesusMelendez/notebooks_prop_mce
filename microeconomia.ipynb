```python

import matplotlib.pyplot as plt
import numpy as np
```



A continuación se presentan una serie de casos para un consumidor en una economía en la cual puede acceder a una combinación de cestas $(x_{1}, {x_2})$  con precios $(p_{1}, p_{2})$ respectivamente; todo esto de acuerdo a su nivel de ingreso $m$.

La situación inicial es la siguiente:



$p_{1}$ = $2$

$p_{2}$ = $4$

$m$ = $2000$



> - $ Restricción $    $Presupuestal:$   $2x_{1} + 4x_{2}  \leq 2000$

> - $ Recta $    $Presupuestal:$   $2x_{1} + 4x_{2}  = 2000$

> - $ Precio Relativo:$  $-\frac{2}{4}$

> - $Cantidad$ $Máxima$ $x_{2}$ : $\frac{2000}{4}$

> - $Cantidad$ $Máxima$ $x_{1}$ : $\frac{2000}{2}$

Reexpresando la recta presupuestal en terminos de $x_{2}$ obtenemos la ecuación de la recta.

$2x_{1} + 4x_{2}$  = $2000$

$4x_{2}  = 2000$ - $2x_{1}$

Resultando 

\begin{equation} x_{2}  = \frac{2000}{4} - \frac{2}{4} x_{1}\end{equation}

```python
# Caso original
precio_1_original = 2

precio_2_original = 4

m_ingreso_original = 2000



# precio relativo original
p_relativo = precio_1_original/precio_2_original

# cantidad máxima x2 original 
q_max_x2 = m_ingreso_original/precio_2_original

# cantidad máxima x1 original
q_max_x1 = m_ingreso_original/precio_1_original

# generamos 100 valores para x1 en el rango de 0 a cantidad máxima de x1 (punto origen abcisa)
x1 = np.linspace(0, q_max_x1, 100)  
x2 = q_max_x2 - p_relativo * x1  #esta es la ecuación de la recta presupuestal 

# Crear el gráfico
# ejes hor ,ver
plt.plot(x1, x2, label=f'$x_{2} = {q_max_x2} - {p_relativo}$')
plt.xlabel('x1')  # Etiqueta del eje x
plt.ylabel('x2')  # Etiqueta del eje y
plt.title('Recta Presupuestal')  # Título del gráfico
plt.grid(True)  # Habilitar la cuadrícula
plt.legend()  # Mostrar la leyenda

# Mostrar el gráfico
plt.show()
```

El caso original (recta azul) podemos observar que se pueden obtener una cantidad máxima de 500 unidades del bien $x_{2}$ o hasta 1000 unidades del bien $x_{1}$.

Como puede notarse, la recta tiene una pendiente negativa, que representa el costo de oportunidad de los bienes $x_{1}$ y $x_{2}$ respectivamente.

```python
#creamos una función para agregar los casos 3-7

def recta_presupuestal(precio_uno=2,precio_dos=4,ingresos=2000,title="Recta Presupuestal",alpha=1):
    # Caso original
    precio_1_original = 2

    precio_2_original = 4

    m_ingreso_original = 2000

    # precio relativo original
    p_relativo = precio_1_original/precio_2_original

    # cantidad máxima x2 original 
    q_max_x2 = m_ingreso_original/precio_2_original

    # cantidad máxima x1 original
    q_max_x1 = m_ingreso_original/precio_1_original

    #precio relativo de caso n
    p_relativo_caso = precio_uno/precio_dos
    #cantidad máxima x2 caso n
    q_max_x2_caso = ingresos/precio_dos
    #cantidad máxima x1 caso n
    q_max_x1_caso = ingresos/precio_uno



    # generamos 100 valores para x1 en el rango de 0 a cantidad máxima de x1 (punto origen abcisa)
    x1 = np.linspace(0, q_max_x1, 100)  
    x2 = q_max_x2 - p_relativo * x1  #esta es la ecuación de la recta presupuestal 

    x1_prima = np.linspace(0,q_max_x1_caso,100)
    x2_prima = q_max_x2_caso - p_relativo_caso * x1_prima

    # Crear el gráfico
    # ejes hor ,ver
    plt.plot(x1, x2, label=f'$x_{2} = {q_max_x2} - {p_relativo}$')
    plt.plot(x1_prima,x2_prima,label=f'$x_{2} = {format(q_max_x2_caso,".1f")} - {format(p_relativo_caso,".1f")}$',color='#ffdf00',alpha=alpha)
    print(f'La nuevas cantidades máximas a aspirar son:\nbien x2: {format(q_max_x2_caso,".1f")}\nx1 es de: {format(q_max_x1_caso,".1f")}')
    plt.xlabel('x1')
    plt.ylabel('x2')  

    plt.title(title) 
    plt.grid(True)  # Habilitar la cuadrícula
    plt.legend() 
  
    return plt.show()
```

#### Recta Presupuestal Caso 3

Se han presentado fluctuaciones en la economía lo cual cambia el precio del bien $x_{2}$.
Situandolo de  4 a 5; el ingreso ($m$) y precio del bien uno ($p_{1}$) permanecen constantes.



$p_{1}$ = $2$

$p_{2}$ = $5$

$m = 2000$

#### Repercusiones


> - $ Restricción $    $Presupuestal:$   $2x_{1} + 5x_{2}  \leq 2000$

> - $ Recta $    $Presupuestal:$   $2x_{1} + 5x_{2}  = 2000$

> - $ Precio Relativo:$  $ - \frac{2}{5}$

> - $Cantidad$ $Máxima$ $x_{2}$ : $\frac{2000}{5}$

> - $Cantidad$ $Máxima$ $x_{1}$ : $\frac{2000}{2}$

```python
recta_presupuestal(precio_uno=2,precio_dos=5,ingresos=2000,title="Recta Presupuestal caso 3")
```

La nueva recta (color amarillo) cambia en el punto de origen de las ordenadas y con una nueva pendiente que tiende al eje de las abcisas respecto al caso original; lo cual indica una **reducción** en la capacidad máxima del consumidor del bien $x_{2}$ pero tambien afecta al reducir el número de canasta de consumo que podía adquirir; esto se refleja en la brecha entre las rectas.
